# House Age vs House Price per Unit

```python
#Scatter Plot between House age, Distance to nearest station and Number of convinience stores 
plt.scatter(x = df["X2 house age"],y = df["Y house price of unit area"],color="green")
plt.title("Relationship Between House Age and House price")
plt.xlabel("House Age")
plt.ylabel("House Prices")
```

### Data Splitting

```python
df.shape
```

```python
training_set = df[:int(len(df)*0.8)]
training_set.shape
```

```python
training_set.tail()
```

```python
testing_set = df[(int(len(df)*0.8)):]
testing_set.shape
```

### Bulding and  Training Model

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
#independend variable 
train_x = np.array(training_set[["X2 house age"]])#independent variable
train_y = np.array(training_set[["Y house price of unit area"]])#dependent variable
```

convert the dataset into array because of viability to machine learnifn models 

```python
#to train dataset, we use fit()
model.fit(train_x,train_y)#fitting model into the training data, 
```

fit(), helps to calculate the line of best fit by finding the coefficient and the Intercept

```python
print ("Coefficient", model.coef_)  # calculating the coeficient

print ("Intercept", model.intercept_) # calculating the intercept
```

    Slope : -0.2858-For each additional year in House age, the price per unit area decreases by approximately 0.28.
    Intercept: 43.00, The predicted pricer per unit arae where House age is 0 is 42

### Y = M*X+C

    y=depended variable/Target variable
    x=independent variable
    m = Coefficient/Slope, where line interepts the y axis
    c=Intercept

```python
plt.scatter(x=df["X2 house age"],y=df["Y house price of unit area"],color="green")
plt.plot(train_x, model.coef_*train_x+model.intercept_,color="r")
plt.xlabel ("House Age")
plt.ylabel("price of House ")
plt.title("Model of House Age  vs House Price")
plt.show()
```

#### Making predictions

```python
def get_regression_method(input_feature, intercept, slope):
    predicted_values = input_feature*slope+intercept
    return predicted_values
```

```python
my_house_age = 32
estimated_price = get_regression_method(my_house_age,model.coef_,model.intercept_)
estimated_price
```

### Model Evaluation

```python
from sklearn.metrics import r2_score
test_x=np.array(testing_set[["X2 house age"]])
test_y=np.array(testing_set[["Y house price of unit area"]])
test_y_ = model.predict(test_x)
```

```python
print("R2 score: ",r2_score(test_y, test_y_))
```

# House Price VS Distance to Nearest MRT station

```python
price_station = df[["X3 distance to the nearest MRT station","Y house price of unit area"]]
price_station.head()


```

```python
price_station.hist()
```

```python
sns.scatterplot(data=price_station,x="X3 distance to the nearest MRT station",y="Y house price of unit area")
plt.title("Scatter plot for Distance to MRT station vs House Price")
```

### Data Splitting

```python
training = price_station[ :int(len(price_station)*0.8)]
training.shape
```

```python
testing = price_station[(int(len(price_station)*0.8)):]
testing.shape
```

### Bulding and Training the Model

```python
from sklearn import linear_model
```

```python
regr  = linear_model.LinearRegression()
training_x = np.array(training[["X3 distance to the nearest MRT station"]])#independent variable
training_y = np.array(training[["Y house price of unit area"]])
regr.fit(training_x,training_y)#ffiting linear Regression model to training data 
print("Coefficient",regr.coef_)
print("Intercept",regr.intercept_)
                      
```

    Slope : -0.0072,For every increase in meters from the nearest MRT station, the house price decreases by 0.0072 pricer per unit area. 
    Intercept: 45.00, The predicted House price per unit area is 45, where the distance to nearest MRT station is 0 meters

### Y = m*X+c

```python
sns.scatterplot(data=price_station,x="X3 distance to the nearest MRT station",y="Y house price of unit area")
plt.plot(training_x,regr.coef_*training_x+regr.intercept_,color="r")#y=mx+c
plt.title("Linear Regression plot for Distance to Nearest station Vs House Price ")
```

### Making Prediction

```python
def get_regression_method(input_feature,intercept,slope):
    predicted_values=input_feature*slope+intercept
    return predicted_values
distance_MRT_station = 10
estimated_price = get_regression_method(distance_MRT_station,regr.coef_,regr.intercept_)
estimated_price
```

### Model Evaluation

```python
testing_x=np.array(testing[["X3 distance to the nearest MRT station"]])
testing_y=np.array(testing[["Y house price of unit area"]])
testing_y_=regr.predict(testing_x)
print(r2_score(testing_y,testing_y_))
```

Model expains 54% of the variance

### Data Splitting

```python
train_dataset= price_stores[: int(len(price_stores)*0.8)]
test_dataset = price_stores[(int(len(price_stores)*0.8)):]
```

### Bulding and Training Model 

```python
from sklearn import linear_model
regression_model=linear_model.LinearRegression()
train_dataset_x = np.array(train_dataset[["X4 number of convenience stores"]])
train_dataset_y = np.array(train_dataset[["Y house price of unit area"]])
regression_model.fit(train_dataset_x,train_dataset_y)
print("Coefficient",regression_model.coef_)
print("Intercept",regression_model.intercept_)
```

### y= mx+c

```python
sns.scatterplot(data=price_stores, x="X4 number of convenience stores",y="Y house price of unit area")
plt.plot(train_dataset_x,regression_model.coef_*train_dataset_x+regression_model.intercept_,color="r")
plt.title("Linear Regression Model for Number of Stores and House price")
```

### Making predictions

```python
def get_regression_method(input_feature, intercept, slope):
    predicted_value = input_feature*slope+intercept
    return predicted_value
number_stores = 2
estimated_price = get_regression_method(number_stores,regression_model.coef_,regression_model.intercept_)
estimated_price


```

### Model Evaluation

```python
from sklearn.metrics import r2_score
test_dataset_x = np.array(test_dataset[["X4 number of convenience stores"]])
test_dataset_y = np.array(test_dataset[["Y house price of unit area"]])
test_y_=regression_model.predict(test_dataset_x)
```

```python
print(r2_score(test_dataset_y, test_y_))
```

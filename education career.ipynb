# Predictive Modelling

#### features selection

```python
num.corr()
```

Predicting Starting salary using features like Gender, Age, Internships completed, projects completed, certifications, soft skills  and filed of study 

```python
#Gender categorical data into binary
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df["Gender"] = le.fit_transform(df["Gender"])
df["Field_of_Study"]=le.fit_transform(df["Field_of_Study"])
df["Current_Job_Level"]=le.fit_transform(df["Current_Job_Level"])
```

```python
df.info()
```

```python
df.head()
```

```python
#features selection
X = df[["Age","Field_of_Study","Gender","Internships_Completed","Projects_Completed","Certifications","Soft_Skills_Score","Current_Job_Level"]]
```

```python
X.head()
```

```python
y=df["Starting_Salary"]
y
```

```python
X.shape
```

### Splitting Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7)
```

### Building and Training Model

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

##### Training Dataset

```python
model.fit(X_train,y_train)#fitting training data
```

```python
print ("Coefficient", model.coef_)  # calculating the coeficient

print ("Intercept", model.intercept_) # calculating the intercept
```

```python
pred=model.predict(X_test)
pred
```

```python
my_data = np.array([[24, 0, 1, 3, 7, 2, 9, 0]])
predictions = model.predict(my_data)
print(predictions)
```

### Model Evaluation

```python
from sklearn.metrics import r2_score
test_y_=model.predict(X_test)
```

```python
print("R2 score: ", r2_score(y_test, test_y_)*100)
```
